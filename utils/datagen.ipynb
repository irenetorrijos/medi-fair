```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import norm, expon, gamma, poisson, uniform, lognorm
```

```python
distributions = {
    'stdnormal': norm(0, 1),
    'exponential': expon(0, 1),
    'gamma': gamma(2, 0, 1),
    'poisson': poisson(1),
    'uniform': uniform(0, 1),
    'lognormal': lognorm(1, 0, 1)
}

child_functions = {
    'linear': lambda x: x*0.5,
    'quadratic': lambda x: x**2,
    'cubic': lambda x: x**3,
    'sinusoidal': lambda x: np.sin(x),
    'exponential': lambda x: np.exp(x),
    'logarithmic': lambda x: np.log(x)
}

noise_functions = {
    'normal': lambda x: norm(0, 1).rvs(x.shape[0]),
    'uniform': lambda x: uniform(0, 1).rvs(x.shape[0])
}
```

# Generating pairs

```python
def generate_pairs(N):
    df = pd.DataFrame(columns=['X', 'Y', 'Xdist', 'childfunc', 'noisefunc'])
    for i in range(N):
        n = random.randint(500, 2000)
        Xdist = random.choice(list(distributions.keys()))
        childfunc = random.choice(list(child_functions.keys()))
        noisefunc = random.choice(list(noise_functions.keys()))

        X = distributions[Xdist].rvs(n)
        Y = child_functions[childfunc](X) + noise_functions[noisefunc](X)

        df = df.append({'X': X, 'Y': Y, 'Xdist': Xdist, 'childfunc': childfunc, 'noisefunc': noisefunc}, ignore_index=True)
    return df

```

```python
data = generate_pairs(100)
```
