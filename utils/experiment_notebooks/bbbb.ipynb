```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generate temperature data
x = np.linspace(0, 12, 100)
temperature = np.sin(x) + np.random.normal(0, 0.1, 100)
```

```python
sunburns = 0.3*temperature**3 + 0.5*temperature + 2*np.random.normal(365)
```

```python
ice_cream_sales = x*temperature + 0.2*np.random.normal(365)
```

```python
sns.pairplot(pd.DataFrame({'temperature': temperature, 'sunburns': sunburns, 'ice_cream_sales': ice_cream_sales}))
```

```python
# normalize
temperature = (temperature - temperature.mean()) / temperature.std()
sunburns = (sunburns - sunburns.mean()) / sunburns.std()
ice_cream_sales = (ice_cream_sales - ice_cream_sales.mean()) / ice_cream_sales.std()

# plot the data
plt.plot(x, sunburns, label='sunburns', alpha=0.5)
plt.plot(x, ice_cream_sales, label='ice cream sales', alpha=0.5)
plt.plot(x, temperature, label='temperature')
plt.title('Normalized sunburns, ice cream sales and temperature plot')
plt.legend()

```

```python
# quantile-quantile plot between temperature and sunburns
plt.figure()
plt.scatter(np.sort(temperature), np.sort(sunburns))
plt.xlabel('temperature')
plt.ylabel('sunburns')

```
