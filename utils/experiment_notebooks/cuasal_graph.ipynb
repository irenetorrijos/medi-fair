```python
import numpy as np
import networkx as nx
```

```python
# generate a random number of nodes
num_nodes = np.random.choice(range(5,21))

# create empty graph
graph = nx.DiGraph()

# add nodes to the graph
graph.add_nodes_from(range(num_nodes))

# number of edges
num_edges = np.random.randint(num_nodes, num_nodes*(num_nodes-1)//3)

# add edges
for i in range(num_edges):
    # random nodes
    source = np.random.randint(0, num_nodes-1)
    target = np.random.randint(0, num_nodes-1)

    # add edge (if not circle)
    if not nx.has_path(graph, target, source):
        graph.add_edge(source, target)


value = nx.is_directed_acyclic_graph(graph)
if value is True:
    # draw the DAG
    nx.draw(graph, with_labels=True)

else:
    print('Generate another one')

```
