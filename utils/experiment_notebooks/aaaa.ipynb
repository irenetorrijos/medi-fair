```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cdt
```

```python
pair1 = cdt.data.load_dataset('tuebingen')[0].iloc[0]
# make df with 2 columns
pair1 = pd.DataFrame({'A': pair1[0], 'B': pair1[1]})
pair1
```

```python
pair1 = pair1[pair1.B > -2]
```

```python
sns.jointplot(x='A', y='B', data=pair1)
```

```python
sns.lmplot(x='B', y='A', data=pair1, fit_reg=True, line_kws={'color': 'red'}, order=2, scatter_kws={'s': 10})
```

```python
sns.lmplot(x='Y', y='X', data=df, fit_reg=True, order=2, line_kws={'color': 'red'})
```

```python
X = np.random.uniform(0, 3, 300)
Y = np.random.uniform(0, 3, 300)
Z = 2*X**2 + Y**2 #+ np.random.randn(300)/3
```

```python
# 2 dimensional gaussian distribution
X = np.random.normal(0, 1, 500)
Y = np.random.normal(0, 1, 500)

X2 = np.random.normal(5, 1, 500)
Y2 = np.random.normal(5, 1, 500)

X3 = np.random.normal(4, 1.5, 500)
Y3 = np.random.normal(1, 1.5, 500)

# 4th blob but with far outliers
X4 = np.random.normal(0, 2, 500)
Y4 = np.random.normal(3, 1, 500)


sns.scatterplot(x=X, y=Y)
sns.scatterplot(x=X2, y=Y2, color='red')
sns.scatterplot(x=X3, y=Y3, color='green')
sns.scatterplot(x=X4, y=Y4, color='yellow')
```

```python
sns.pairplot(pd.DataFrame({'X': X, 'Y': Y, 'Z': Z}))
```

```python
# Plot X and Y using seaborn and add labels on the axes
sns.scatterplot(x=X, y=Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

sns.scatterplot(x=Y, y=X)
plt.xlabel('Y')
plt.ylabel('X')
plt.show()
```

```python
from cdt.causality.pairwise import CDS

# Create an instance of the ANM class
anm = CDS()
print(anm.predict_proba((X.reshape(-1, 1), Y.reshape(-1, 1))))
print(anm.predict_proba((Y.reshape(-1, 1), X.reshape(-1, 1))))
```
